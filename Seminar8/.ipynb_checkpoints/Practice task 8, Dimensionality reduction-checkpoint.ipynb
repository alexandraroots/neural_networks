```python
def plot_dataset(X, emb, margin=50, frame=2000., imsize=64):
    assert(emb.shape[1] == 2)
    emb = np.array((emb - emb.min(0)) * (frame / (emb.max(0) - emb.min(0))) + margin, dtype = np.int)
    image = np.zeros((int(emb[:, 0].max() + margin + imsize), int(emb[:, 1].max() + margin + imsize))) + 1
    for k in range(X.shape[0]):
        image[emb[k, 0]:emb[k, 0]+64, emb[k, 1]:emb[k, 1]+64] = X[k].reshape(64, 64)
    plt.figure(figsize=(20, 20))
    
    plt.imshow(image, cmap="Greys_r")
    plt.axis("off")
    plt.show()
```

###  Часть 2: Автокодировщик

<i> 2.1 Обучите автокодировщик с архитектурой 4096 -> 200 -> 10 -> 200 -> 4096. В качестве нелинейностей используйте сигмоиду.</i>

```python
from torch.utils.data import DataLoader, Dataset
```

```python
dataloader = DataLoader(X, batch_size=400, shuffle=True)
```

```python
def fully_connected(arc, last_nonlinearity=True):
    layers = []
    for i in range(len(arc)-1):
        layers.append(nn.Linear(arc[i], arc[i+1]))
        if i != len(arc)-2 or last_nonlinearity:
            layers.append(nn.Sigmoid())
    return nn.Sequential(*layers)
```

```python
fully_connected([4096, 200, 10], last_nonlinearity=False)
```

```python
class AE(nn.Module):
    def __init__(self, arc_encoder, last_nonlinearity_dec, arc_decoder, last_nonlinearity_enc):
        super(AE, self).__init__()
        self.encoder = fully_connected(arc_encoder, last_nonlinearity_enc)
        self.decoder = fully_connected(arc_decoder, last_nonlinearity_dec)
        
    def forward(self, X):
        return self.decoder(self.encoder(X))
```

```python
model = AE([4096, 200], False, [200, 4096], False)
```

```python
model
```

```python
loss = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import sys
```

```python
history_loss = []
for epoch in range(200):
    losses = []
    for x in dataloader:
        x_batch = Variable(x)
        y_batch = Variable(x, requires_grad=False)
        model.zero_grad()
        x_reconstruction = model(x_batch)
        loss_batch = loss(x_reconstruction, y_batch)
        loss_batch.backward()
        optimizer.step()
        losses.append(loss_batch.data.item())
    history_loss.append(np.mean(losses))
    sys.stdout.write('\r {}'.format(np.mean(losses)))
```

```python
embedding = model.encoder(Variable(torch.Tensor(X))).data.numpy()
```

```python
reconstruction = model(Variable(torch.Tensor(X))).data.numpy()
```

<i> 2.3 Обучите сеть с разными размерностями пространства, в которое происходит отображение </i>

```python
latent_spaces = np.append(np.arange(1, 10), np.round(np.logspace(1, np.log10(200), 5)).astype(int))
latent_spaces
```

```python
from torch.utils.data import DataLoader, Dataset
X_Tensor = torch.FloatTensor(torch.from_numpy(X))
X_Tensor -= X_Tensor.mean(0, keepdim=True)
train_loader = DataLoader(X_Tensor, 400, shuffle=True)
```

```python
networks = []
for latent_shape in latent_spaces:
    networks.append(AE([4096, 200, latent_shape], True, [latent_shape, 200, 4096], False))
```

```python
def train(network, epochs, learning_rate):
    loss = nn.MSELoss().cuda()
    train_loss_epochs = []
    test_loss_epochs = []
    train_accuracy_epochs = []
    test_accuracy_epochs = []
    optimizer = torch.optim.Adam(network.parameters(), lr=learning_rate)
    try:
        for epoch in range(epochs):
            losses = []
            for X in dataloader:
                X = Variable(X)
                y = X
                network.zero_grad()
                prediction = network(X)
                loss_batch = loss(prediction, y)
                losses.append(loss_batch.item())
                loss_batch.backward()
                optimizer.step()
            train_loss_epochs.append(np.mean(losses))
            losses = []
            accuracies = []    
            for X in dataloader:
                X = Variable(X)
                y = X
                prediction = network(X)
                loss_batch = loss(prediction, y)
                losses.append(loss_batch.item())
            test_loss_epochs.append(np.mean(losses))
            sys.stdout.write('\rEpoch {0}... (Train/Test) MSE_LOSS: {1:.3f}/{2:.3f}\t'.format(
                        epoch, train_loss_epochs[-1], test_loss_epochs[-1]))
    except KeyboardInterrupt:
        pass
```

```python
evrs = np.zeros(len(networks))

X_Variable = Variable(X_Tensor, requires_grad=False)
X_Variable -= X_Variable.mean(0, keepdim=True)
for idx, latent_space in enumerate(latent_spaces):
    print (latent_space)
    for _ in range(20):
        train(networks[idx], 100, 0.01)
        y = networks[idx](X_Variable)
        err2 = ((y - X_Variable) ** 2).sum(1).mean()
        evr = 1 - err2 / (X_Variable ** 2).sum(1).mean()
        #print("Dor", evr.data.cpu().numpy())
        evr = evr.data.cpu().numpy()
        evrs[idx] = max(evr, evrs[idx])
```

<i> 2.4 Сравните графики из пунктов 1.2 и 2.3 </i>

```python
model = PCA()
model.fit(X)

plt.plot(latent_spaces, evrs, label='AE')
plt.plot(np.cumsum(model.explained_variance_ratio_)[:200], label='PCA')
plt.legend()
```
